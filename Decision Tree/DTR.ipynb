```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/srinivasav22/Graduate-Admission-Prediction/master/Admission_Predict_Ver1.1.csv")
df
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.duplicated().sum()
```

```python
X = df.drop("Chance of Admit ",axis=1)
y = df["Chance of Admit "]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.35,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(DecisionTreeRegressor(),{
    "criterion":["squared_error", "friedman_mse", "absolute_error", "poisson"],
    "max_depth" : range(2,32,1),
    "min_samples_leaf" : range(2,10,1),
    "splitter":["best","random"]
},cv=5,return_train_score=False)

clf.fit(X_train,y_train)
clf.cv_results_
```

```python
clf.best_params_
```

```python
model2 = DecisionTreeRegressor(criterion= 'friedman_mse',
 max_depth= 16,
 min_samples_leaf= 8,
 splitter= 'random')
```

```python
model2.fit(X_train,y_train)
```

```python
ypred2 = model2.predict(X_test)
```

```python
r2_score(y_test,ypred2)
```
